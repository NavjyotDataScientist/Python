```python
# Project 4: Simple Text Analyzer
# This project helps beginners understand strings and text analysis in Python

# Step 1: Take input from the user
text = input("Enter a sentence: ")

# Step 2: Initialize counters
num_chars = len(text)
words = text.split()
num_words = len(words)
vowels = "aeiouAEIOU"
num_vowels = 0
num_consonants = 0

# Step 3: Count vowels and consonants
for char in text:
    if char.isalpha():
        if char in vowels:
            num_vowels += 1
        else:
            num_consonants += 1

# Step 4: Display results
print("\nText Analysis")
print("-------------")
print(f"Sentence: {text}")
print(f"Total Characters: {num_chars}")
print(f"Total Words: {num_words}")
print(f"Vowels: {num_vowels}")
print(f"Consonants: {num_consonants}")
```
