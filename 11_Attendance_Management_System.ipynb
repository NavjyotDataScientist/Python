Goal (Student Understanding)

Create a simple system to:

mark student attendance

store data in a dictionary

display present & absent students

This introduces structured data, like a real dataset.

```python
attendance = {
    "Rahul": "Present",
    "Priya": "Absent",
    "Amit": "Present",
    "Sneha": "Present",
    "Karan": "Absent"
}

present_students = []
absent_students = []

for student, status in attendance.items():
    if status == "Present":
        present_students.append(student)
    else:
        absent_students.append(student)

print("ðŸ“‹ Attendance Report")
print("Present Students:", present_students)
print("Absent Students:", absent_students)
```
