```python
# Project 6: Student Performance & Analytics Tracker

# Step 1: Create an empty dictionary to store student data
students = {}

# Step 2: Input student data
num_students = int(input("Enter number of students: "))

for _ in range(num_students):
    name = input("\nEnter student name: ")
    num_subjects = int(input(f"How many subjects for {name}? "))

    marks = []
    for i in range(num_subjects):
        mark = float(input(f"Enter mark for subject {i+1}: "))
        marks.append(mark)

    students[name] = marks

# Step 3: Calculate averages and print report
print("\n--- Student Report ---")
for student, marks in students.items():
    total = sum(marks)
    average = total / len(marks)
    print(f"{student}: Total = {total}, Average = {average:.2f}")

# Step 4: Calculate class average per subject
print("\n--- Class Average per Subject ---")
num_subjects = max(len(marks) for marks in students.values())
for i in range(num_subjects):
    subject_total = 0
    count = 0
    for marks in students.values():
        if i < len(marks):
            subject_total += marks[i]
            count += 1
    subject_avg = subject_total / count
    print(f"Subject {i+1} Average: {subject_avg:.2f}")

# Step 5: Find top-performing student
top_student = max(students, key=lambda x: sum(students[x])/len(students[x]))
top_avg = sum(students[top_student])/len(students[top_student])
print(f"\nTop Student: {top_student} with average {top_avg:.2f}")
```
