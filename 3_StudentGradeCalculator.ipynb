```python
# Project 3: Student Grade Calculator
# This project helps beginners understand conditional logic in Python

print("Student Grade Calculator")

# Step 1: Take input from user
marks = int(input("Enter your marks (0-100): "))

# Step 2: Calculate grade using if-elif-else
if marks >= 90 and marks <= 100:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 50:
    grade = "D"
elif marks >= 0:
    grade = "F"
else:
    grade = "Invalid marks"

# Step 3: Display the result
print(f"Marks: {marks}")
print(f"Grade: {grade}")
```
