```python
# Project 5: Mini CSV Expense Tracker
# This project introduces Pandas and basic data analysis

import pandas as pd

# Step 1: Load data from CSV
data = pd.read_csv("expenses.csv")

# Step 2: Analyze the expenses
total_expense = data["Expense"].sum()
average_expense = data["Expense"].mean()
highest_expense = data["Expense"].max()
lowest_expense = data["Expense"].min()

# Step 3: Display results
print("Mini CSV Expense Tracker")
print("------------------------")
print(data)
print(f"\nTotal Expense: {total_expense}")
print(f"Average Daily Expense: {average_expense}")
print(f"Highest Expense: {highest_expense}")
print(f"Lowest Expense: {lowest_expense}")
```
