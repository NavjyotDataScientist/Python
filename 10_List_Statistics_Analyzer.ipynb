goal (Student Mindset)

Analyze a list of numbers and calculate:

total

minimum

maximum

average

This is the first real step toward EDA (Exploratory Data Analysis).

```python
numbers = [45, 67, 89, 23, 56, 78, 90, 34]

total = sum(numbers)
minimum = min(numbers)
maximum = max(numbers)
average = total / len(numbers)

print("Numbers:", numbers)
print("Total:", total)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("Average:", round(average, 2))
```
