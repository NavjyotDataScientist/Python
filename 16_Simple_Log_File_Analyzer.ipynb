Goal (Student Understanding)

Analyze a log file and count:

total log entries

ERROR messages

INFO messages

This is industry-style data analysis.

```python
log_data = [
    "INFO User logged in",
    "ERROR Failed to load page",
    "INFO Data saved successfully",
    "ERROR Database connection failed",
    "INFO User logged out"
]

error_count = 0
info_count = 0

for log in log_data:
    if log.startswith("ERROR"):
        error_count += 1
    elif log.startswith("INFO"):
        info_count += 1

print("ðŸ“Š Log Analysis Report")
print("Total Logs:", len(log_data))
print("ERROR Logs:", error_count)
print("INFO Logs:", info_count)
```
