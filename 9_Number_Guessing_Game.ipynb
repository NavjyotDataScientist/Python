Goal (Student Understanding)

Create a simple game where:

computer selects a random number

user keeps guessing

program gives hints until correct

This builds:

loops

conditions

random module

logical flow (very important)

ðŸ§  Game Rules

Number range: 1 to 100

User keeps guessing until correct

Show hints:

Too High

Too Low

Count attempts

```python
import random

# Computer selects a random number
secret_number = random.randint(1, 100)

attempts = 0
guess = None

print("ðŸŽ¯ Welcome to the Number Guessing Game!")
print("Guess a number between 1 and 100")

while guess != secret_number:
    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess > secret_number:
        print("Too High! Try again.")
    elif guess < secret_number:
        print("Too Low! Try again.")
    else:
        print(f"ðŸŽ‰ Correct! You guessed it in {attempts} attempts.")
```
