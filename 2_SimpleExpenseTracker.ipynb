```python
# Project 2: Simple Expense Tracker
# This project helps beginners understand how real-world data is collected and analyzed

expenses = []

print("Simple Expense Tracker")
print("Enter your expenses for 5 days")

# Step 1: Take input from the user
for day in range(1, 6):
    amount = float(input(f"Enter expense for day {day}: "))
    expenses.append(amount)

# Step 2: Perform analysis
total_expense = sum(expenses)
average_expense = total_expense / len(expenses)
highest_expense = max(expenses)
lowest_expense = min(expenses)

# Step 3: Display results
print("\nExpense Summary")
print("----------------")
print("Expenses:", expenses)
print("Total Expense:", total_expense)
print("Average Daily Expense:", average_expense)
print("Highest Expense:", highest_expense)
print("Lowest Expense:", lowest_expense)
```
