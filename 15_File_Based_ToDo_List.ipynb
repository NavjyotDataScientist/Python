Goal (Student Understanding)

Create a simple To-Do List that:

saves tasks to a file

reads tasks back from the file

This teaches data persistence.

ðŸ§  Core Idea

Data should not disappear when the program ends.
This is how real applications work.

```python
filename = "tasks.txt"

def add_task(task):
    with open(filename, "a") as file:
        file.write(task + "\n")

def view_tasks():
    with open(filename, "r") as file:
        tasks = file.readlines()
        print("ðŸ“‹ Your Tasks:")
        for task in tasks:
            print("-", task.strip())

add_task("Learn Python Basics")
add_task("Practice Data Science Problems")
view_tasks()
```
