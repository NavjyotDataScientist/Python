Goal (Student Understanding)

Count how many times each word appears in a sentence.

This is the FIRST STEP toward NLP (Natural Language Processing).

```python
text = "data science is fun and data science is powerful"

words = text.lower().split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Word Frequency:")
print(word_count)
```
