```python
# Project 1: Student Marks Analyzer
# This project helps beginners understand how data analysis starts in Python

# Step 1: Store marks in a list
marks = [78, 85, 92, 67, 88]

# Step 2: Calculate total marks
total_marks = sum(marks)

# Step 3: Calculate average marks
average_marks = total_marks / len(marks)

# Step 4: Find highest and lowest marks
highest_marks = max(marks)
lowest_marks = min(marks)

# Step 5: Display the results
print("Student Marks Analysis")
print("----------------------")
print("Marks:", marks)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Marks:", highest_marks)
print("Lowest Marks:", lowest_marks)
```
